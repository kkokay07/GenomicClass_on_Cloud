# **DNA Repair**

- **Definition**: DNA repair encompasses cellular mechanisms that identify and correct damage to DNA, maintaining genomic stability.  

- **Types of DNA Repair Mechanisms**:  
  - **Direct Repair**: Fixes specific lesions like alkylation damage.  
  - **Base Excision Repair (BER)**: Repairs small, non-helix-distorting base lesions.  
  - **Nucleotide Excision Repair (NER)**: Removes bulky, helix-distorting lesions.  
  - **Mismatch Repair (MMR)**: Corrects replication errors such as mismatches and insertion-deletion loops.  
  - **Homologous Recombination (HR)** and **Non-Homologous End Joining (NHEJ)**: Repair double-strand breaks.  
  - **Translesion Synthesis (TLS)**: Bypasses lesions during replication, albeit error-prone.  

- **Importance**:  
  - **Genomic Integrity**: Prevents mutations and chromosomal abnormalities.  
  - **Disease Prevention**: Protects against cancer, neurodegenerative diseases, and premature aging.  
  - **Cell Survival**: Maintains viability under conditions of DNA damage caused by UV light, radiation, or chemicals.  
  - **Reproductive Fitness**: Essential for accurate DNA transmission during replication and meiosis.  
  - **Therapeutic Relevance**: Targeted manipulation of DNA repair pathways aids in cancer treatment and gene editing.  

# **I. Direct repair of DNA**

1. **Definition:** Direct repair is a DNA repair mechanism that directly reverses damage to DNA without requiring the removal of damaged nucleotides or the synthesis of new DNA.

2. **Types of Damage Repaired:** It primarily addresses specific lesions like alkylation, UV-induced cyclobutane pyrimidine dimers, and certain oxidative damages.

3. **Key Enzymes:**

    O6-Methylguanine-DNA methyltransferase (MGMT): Removes alkyl groups from the O6 position of guanine.
    Photolyase: Utilizes light energy to repair UV-induced pyrimidine dimers (common in prokaryotes and some eukaryotes).

4. **Energy Requirements: **Direct repair mechanisms are generally energy-efficient compared to other DNA repair processes, as they do not involve extensive processing or synthesis steps.

5. **Error-Free: **It is considered an error-free repair process because it restores the original DNA sequence without introducing mutations.
Significance: This repair mechanism helps maintain genomic integrity and prevent mutagenesis or cell death due to DNA damage.

# **Base Excision Repair**

Base Excision Repair (BER) is a DNA repair mechanism responsible for correcting small, non-helix-distorting base lesions, such as those caused by oxidation, deamination, or alkylation.

The process involves:

    Recognition and removal of the damaged base by a DNA glycosylase, leaving an abasic (AP) site.
    Cleavage of the sugar-phosphate backbone at the AP site by an AP endonuclease.
    Removal of the sugar residue by a lyase or phosphodiesterase.
    Synthesis of a new nucleotide by DNA polymerase, using the complementary strand as a template.
    Sealing of the nick in the DNA backbone by DNA ligase, restoring the DNA's integrity.

This pathway is crucial for maintaining genomic stability and preventing mutations that could lead to diseases such as cancer.

Following figure description
*  Certain forms of base damage are recognized by DNA glycosylases
*  DNA glycosylases catalyze excision of the free base by hydrolysis of the N-glycosyl bond linking the base to the sugar–phosphate backbone.
*  This reaction leaves an apurinic or apyrimidinic (AP) site in the DNA.
*  Attack at such sites by a 5′ AP endonuclease results in a strand break with a 5′ terminal deoxyribose–phosphate moiety.
*   These are excised by the action of a DNA deoxyribophosphodiesterase (dRpase)
*   The resulting single nucleotide gap is filled by repair synthesis and DNA repair is completed by DNA ligase

# **Nucleotide Excision Repair (NER)**

It is a versatile DNA repair mechanism that removes bulky, helix-distorting lesions such as UV-induced thymine dimers, chemical adducts, and cross-links.  

### Key Steps in NER:  
1. **Damage Recognition**: Specialized proteins detect distortions in the DNA helix caused by lesions.  
   - **Global Genomic NER (GG-NER)**: Recognizes damage throughout the genome.  
   - **Transcription-Coupled NER (TC-NER)**: Focuses on lesions in actively transcribed genes.  

2. **Unwinding of DNA**: Helicases (e.g., XPB and XPD in humans) unwind the DNA around the lesion to form a bubble structure.  

3. **Incision**: Endonucleases cleave the damaged strand on both sides of the lesion, typically 24–32 nucleotides apart.  

4. **Excision**: The damaged oligonucleotide is removed, leaving a gap in the DNA.  

5. **Synthesis and Ligation**: DNA polymerase fills the gap using the undamaged strand as a template, and DNA ligase seals the strand, restoring the DNA.  

### Importance:  
NER is critical for preventing mutations and diseases like xeroderma pigmentosum (XP), which arises from defects in NER pathways, leading to extreme UV sensitivity and increased cancer risk.

# **III. Mismatch Repair**

**Mismatch Repair (MMR)** is a DNA repair mechanism that identifies and corrects base-pair mismatches and insertion-deletion loops (IDLs) that occur during DNA replication.  

### Key Features of MMR:  
1. **Damage Type**: Repairs mismatches that escape proofreading by DNA polymerase, such as incorrect base-pairing (e.g., A-C, G-T) or small insertion/deletion loops resulting from replication slippage.  

2. **Recognition**:  
   - In bacteria, **MutS** recognizes the mismatch and recruits **MutL**.  
   - In eukaryotes, **MSH2-MSH6** (for single-base mismatches) or **MSH2-MSH3** (for larger IDLs) perform mismatch recognition.  

3. **Strand Discrimination**: The repair machinery distinguishes the newly synthesized strand from the template strand by recognizing nicks (in eukaryotes) or hemimethylation of GATC sites (in bacteria).  

4. **Excision**: An exonuclease removes a segment of the newly synthesized strand containing the mismatch.  

5. **Resynthesis and Ligation**: DNA polymerase fills the gap using the template strand, and DNA ligase seals the nick, restoring the DNA sequence.  

### Significance:  
MMR is crucial for maintaining replication fidelity and genome stability. Defects in MMR genes (e.g., **MLH1**, **MSH2**, **MSH6**, **PMS2**) are associated with **Lynch syndrome**, a hereditary cancer predisposition syndrome.

# IV. Recombinational Repair

**Recombinational Repair** is a DNA repair mechanism that uses homologous sequences to repair double-strand breaks (DSBs) or interstrand cross-links, ensuring the restoration of the DNA's integrity and sequence fidelity.

### Key Features of Recombinational Repair:

1. **Types of Damage Repaired**:
   - Double-strand breaks caused by ionizing radiation, chemical agents, or replication fork collapse.
   - Interstrand cross-links and single-strand gaps during DNA replication.

2. **Pathways**:
   - **Homologous Recombination (HR)**:
     - Requires a homologous sequence (usually the sister chromatid) as a template.
     - Initiated by the resection of DSB ends to generate single-stranded DNA, which invades the homologous DNA, forming a displacement loop (D-loop).
     - DNA synthesis, branch migration, and resolution restore the sequence.
   - **Non-Homologous End Joining (NHEJ)** (error-prone and not strictly recombinational):
     - Direct ligation of DSB ends without the need for homology.
     - More active in G1 phase when no sister chromatid is available.

3. **Key Proteins**:
   - **Rad51/RecA**: Facilitates strand invasion and homology search.
   - **BRCA1 and BRCA2**: Regulate HR and Rad51 activity in eukaryotes.
   - **RuvABC and RecBCD**: Involved in processing and resolving Holliday junctions in prokaryotes.

4. **Cell Cycle Dependence**:
   - HR predominantly occurs during the S and G2 phases when a sister chromatid is available.
   - NHEJ is more common in G1 phase.

5. **Biological Significance**:
   - Prevents chromosomal instability, loss of genetic information, and mutagenesis.
   - HR is crucial for meiosis and maintaining genomic integrity during DNA replication.

6. **Clinical Implications**:
   - Defects in HR (e.g., mutations in **BRCA1/BRCA2**) are linked to cancer predisposition.
   - Targeted inhibition of HR (e.g., PARP inhibitors) is a therapeutic strategy for BRCA-deficient tumors.

# **V. Error-Prone Translesion DNA Repair**

**Error-Prone Translesion DNA Repair** (also known as translesion synthesis, TLS) is a DNA damage tolerance mechanism that allows the DNA replication machinery to bypass lesions that would otherwise stall the replication fork.  

### Key Features of Error-Prone Translesion DNA Repair:

1. **Mechanism**:
   - Specialized low-fidelity DNA polymerases temporarily replace the high-fidelity replicative polymerase.
   - These TLS polymerases synthesize DNA across damaged templates, incorporating nucleotides opposite lesions that would block normal replication.  

2. **Key Enzymes**:
   - **Pol η (eta)**: Handles UV-induced thymine dimers with high accuracy.
   - **Pol κ (kappa), Pol ι (iota), and Pol ζ (zeta)**: Operate on various lesions but are often error-prone, introducing mutations.  

3. **Damage Bypassed**:
   - UV-induced pyrimidine dimers.
   - Abasic sites, bulky adducts, and cross-linked bases caused by chemicals.
   - Oxidative and alkylated base lesions.  

4. **Error-Prone Nature**:
   - TLS polymerases lack proofreading activity and have lower fidelity, leading to increased mutation rates.
   - Mutations introduced can be beneficial (e.g., in adaptive immunity) or harmful (e.g., cancer development).  

5. **Biological Significance**:
   - Provides a survival mechanism under conditions of severe DNA damage.
   - Prevents replication fork collapse and cell death by allowing replication to continue past DNA lesions.  

6. **Clinical Implications**:
   - Overactivation of TLS contributes to mutagenesis and therapy resistance in cancer cells.
   - TLS inhibitors are being explored to enhance the effectiveness of chemotherapies and radiation therapy.  

Error-prone translesion repair balances immediate survival against the risk of increased mutagenesis, reflecting a trade-off between damage tolerance and genomic stability.
