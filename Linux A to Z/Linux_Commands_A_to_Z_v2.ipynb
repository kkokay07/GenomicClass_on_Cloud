## 📁 Navigation & Directory Commands

### `pwd` — Print Working Directory
**What it does:** Shows you exactly where you are in the filesystem right now.
**Syntax:** `pwd`
**Used for:** Knowing your current location before running other commands.

```python
%%bash
# pwd stands for 'Print Working Directory'
# It prints the full path of the folder you are currently inside
# No options needed — just type pwd and press Enter
pwd
```

### `ls` — List Directory Contents
**What it does:** Shows all files and folders inside a directory.
**Syntax:** `ls [options] [directory]`
**Used for:** Seeing what files exist in a folder.

```python
%%bash
# ls — basic listing of files in the current directory
ls /etc | head -8
```

```python
%%bash
# ls -l — 'long format' showing permissions, owner, size, and date
# Each column means: permissions | links | owner | group | size | date | name
ls -l /tmp | head -5
```

```python
%%bash
# ls -a — shows ALL files, including hidden ones (files that start with a dot .)
ls -a ~ | head -8
```

```python
%%bash
# ls -lh — 'long format' with 'human-readable' file sizes (KB, MB, GB)
# Combining -l and -h makes sizes easy to read
ls -lh /etc | head -5
```

### `cd` — Change Directory
**What it does:** Moves you from one folder to another.
**Syntax:** `cd [directory_path]`

## 🔐 Permissions & Ownership

### `chmod` — Change File Permissions
**What it does:** Controls who can read, write, or execute a file.
**Syntax:** `chmod [permissions] filename`
**Used for:** Securing files, making scripts executable.

## **Owner**

* The user who created the file or has been assigned ownership.
* Has primary control over the file.
* Can modify permissions and content depending on allowed rights.

## **Group**

* A collection of users grouped together for shared access.

* All users in that group receive the permissions defined for the group field.

* seful for collaborative environments (e.g., research lab members sharing data).

## **Others**

* All remaining users on the system who are neither the owner nor part of the group.

* Usually given limited permissions for security reasons.

```python
%%bash
# Create a test file
touch /tmp/myscript.sh

# ls -l shows permissions in format: -rwxrwxrwx
# Position 1: file type (- = file, d = directory, l = link)
# Positions 2-4: OWNER permissions (r=read, w=write, x=execute)
# Positions 5-7: GROUP permissions
# Positions 8-10: OTHERS permissions
echo "Default permissions:"
ls -l /tmp/myscript.sh
```

```python
%%bash
# chmod 755 — numeric method:
# 7 = rwx (4+2+1) for owner
# 5 = r-x (4+0+1) for group
# 5 = r-x (4+0+1) for others
chmod 755 /tmp/myscript.sh
echo "After chmod 755:"
ls -l /tmp/myscript.sh
```

```python
%%bash
# chmod +x — symbolic method: adds execute (x) permission for everyone
chmod +x /tmp/myscript.sh
echo "After chmod +x:"
ls -l /tmp/myscript.sh
```

```python
%%bash
# chmod 600 — only owner can read and write; no one else can access
chmod 600 /tmp/myscript.sh
echo "After chmod 600 (private):"
ls -l /tmp/myscript.sh
```

## 📝 Text Processing

### `sort` — Sort Lines of Text
**What it does:** Arranges lines of a file in alphabetical or numerical order.
**Syntax:** `sort [options] filename`
**Used for:** Organizing data, removing duplicates, ranking.

```python
%%bash
# Create a sample file with unsorted data
echo -e "Banana\nApple\nCherry\nDate\nApricot" > /tmp/unsorted.txt
```

```python
%%bash
# sort with no options sorts alphabetically A to Z
sort /tmp/unsorted.txt
```

```python
%%bash
# -k 2 means sort by the second COLUMN (space-separated)
echo -e "Alice 88\nBob 72\nCarol 95" | sort -k2 -n
```

### `uniq` — Remove Duplicate Lines
**What it does:** Filters out repeated adjacent lines from sorted input.
**Syntax:** `uniq [options] filename`
**Used for:** Removing duplicates, counting occurrences.

> ⚠️ Note: `uniq` only removes ADJACENT duplicates — always sort first!

```python
%%bash
# Create a file with duplicates
echo -e "Apple\nApple\nBanana\nBanana\nBanana\nCherry" > /tmp/dupes.txt
```

```python
%%bash
cat /tmp/dupes.txt
```

```python
%%bash
# uniq keeps only ONE copy of each repeated adjacent line
uniq /tmp/dupes.txt
```

```python
%%bash
# -c prepends each line with how many times it appeared
uniq -c /tmp/dupes.txt
```

```python
%%bash
# -d shows only lines that appeared MORE THAN ONCE
uniq -d /tmp/dupes.txt
```

```python
%%bash
# Sort first so all duplicates become adjacent, then uniq removes them
echo -e "Cherry\nApple\nBanana\nApple\nCherry" | sort | uniq
```

## 💾 File Compression & Archiving

### `tar` — Archive Files (Tape Archive)
**What it does:** Bundles multiple files into a single archive file (with optional compression).
**Syntax:** `tar [options] archive_name files_to_archive`
**Used for:** Backups, transferring multiple files as one package.

```python
%%bash
# Create some files to archive
mkdir -p /tmp/tartest
echo "Document 1" > /tmp/tartest/doc1.txt
echo "Document 2" > /tmp/tartest/doc2.txt
echo "Document 3" > /tmp/tartest/doc3.txt
```

```python
%%bash
# CREATE a tar.gz archive
# tar -czf: c=create, z=use gzip compression, f=filename follows
# archive.tar.gz is the output file name
# -C /tmp says 'go into /tmp first' so paths inside archive are clean
tar -czf /tmp/archive.tar.gz  -C /tmp tartest
```

```python
%%bash
ls -lh /tmp/archive.tar.gz
```

```python
%%bash
# LIST contents without extracting
# tar -tzf: t=list contents, z=gzip, f=filename — does NOT extract
tar -tzf /tmp/archive.tar.gz
```

```python
%%bash
# EXTRACT the archive
# tar -xzf: x=extract, z=gzip, f=filename
# -C /tmp/extracted specifies WHERE to put the extracted files
mkdir -p /tmp/extracted
tar -xzf /tmp/archive.tar.gz -C /tmp/extracted
```

```python
%%bash
ls /tmp/extracted/tartest/
```

### `gzip` / `gunzip` — Compress and Decompress Files
**What it does:** Compresses a single file using the gzip algorithm.
**Syntax:** `gzip filename` / `gunzip filename.gz`
**Used for:** Reducing file size for storage or transfer.

```python
%%bash
# Create a test file
echo "This is a file that will be compressed using gzip." > /tmp/bigfile.txt
```

```python
%%bash
# gzip filename — compresses the file and REPLACES it with filename.gz
# The original file is deleted and a .gz file is created
gzip /tmp/bigfile.txt
```

```python
%%bash
# gunzip — decompresses a .gz file and REPLACES it with the original
gunzip /tmp/bigfile.txt.gz
```

```python
%%bash
# gzip -k — keep the original file (don't delete it)
gzip -k /tmp/bigfile.txt
echo "Both files exist:"
ls -lh /tmp/bigfile.txt /tmp/bigfile.txt.gz
```

### `zip` / `unzip` — Create and Extract ZIP Archives
**What it does:** Creates .zip format archives (compatible with Windows/Mac).
**Syntax:** `zip archive.zip files` / `unzip archive.zip`
**Used for:** Cross-platform archiving, sharing with Windows users.

```python
%%bash
# Create files to zip
echo "File A" > /tmp/fileA.txt
echo "File B" > /tmp/fileB.txt

# zip output.zip file1 file2 — creates a ZIP archive with those files inside
cd /tmp && zip myfiles.zip fileA.txt fileB.txt
```

```python
%%bash
# unzip -l — LISTS contents of a zip file without extracting
unzip -l /tmp/myfiles.zip
```

```python
%%bash
# unzip -d — extracts to a specific DESTINATION directory
mkdir -p /tmp/unzipped
unzip -o /tmp/myfiles.zip -d /tmp/unzipped   # -o = overwrite without asking (-o tells, if a file with the same name already exists in the destination folder, replace it automatically.)
echo "Extracted files:"
ls /tmp/unzipped/
```

**Syntax:** `ps [options]`
**Used for:** Monitoring what programs are running, finding PIDs to kill.

```python
%%bash
# ps with no options shows only processes in your current shell session
ps
```

```python
%%bash
# ps aux (ALL processes)
# a = all users, u = user-oriented format, x = include processes without a terminal
# This is the most commonly used combination
# Columns: USER PID %CPU %MEM VSZ RSS TTY STAT START TIME COMMAND
ps aux | head -8
```

Explanation of each column:

* **USER** – Username of the account that started the process.

* **PID** – Process ID; a unique number assigned to each running process.

* **%CPU** – Percentage of CPU resources currently used by the process.

* **%MEM** – Percentage of physical RAM used by the process.

* **VSZ** – Virtual memory size of the process (in kilobytes). Includes memory that may not currently be in RAM.

* **RSS** – Resident Set Size; actual physical RAM used by the process (in kilobytes).

* **TTY** – Terminal associated with the process. `?` means it is not attached to any terminal (background/system process).

* **STAT** – Current status of the process (running, sleeping, stopped, etc.).

* **START** – Time or date when the process started.

* **TIME** – Total CPU time consumed by the process since it started.

* **COMMAND** – The command or program that started the process, including arguments.

Important to note the key difference of:

* CPU % → processing activity (how hard the processor is working).

* RAM % → memory usage (how much data the program occupies in memory).

```python
%%bash
# Find a specific process
# Pipe ps output into grep to search for a specific program
ps aux | grep bash | head -3
```

### `kill` — Terminate a Process
**What it does:** Sends a signal to a process — usually to stop it.
**Syntax:** `kill [signal] PID`
**Used for:** Stopping frozen or unwanted processes.

```python
%%bash
# Start a background process (sleep 100 runs silently for 100 seconds)
# The & at the end runs it in the BACKGROUND
sleep 100 &

# $! captures the PID (Process ID) of the last background command
PID=$!
echo "Started background process with PID: $PID"

# Verify it's running by searching for it
ps aux | grep "sleep 100" | grep -v grep

# kill PID — sends SIGTERM (signal 15) asking the process to stop gracefully
kill $PID
echo "Sent kill signal to PID $PID"

# Verify it's gone
sleep 0.5   # brief wait for the process to die
ps aux | grep "sleep 100" | grep -v grep || echo "Process is gone!"

echo ""
echo "Common Kill Signals:"
echo "  kill PID     = SIGTERM (15) — polite request to stop"
echo "  kill -9 PID  = SIGKILL (9)  — force kill immediately"
echo "  kill -1 PID  = SIGHUP  (1)  — reload config without restarting"
```

## 🧠 Advanced Text Processing

### `sed` — Stream Editor
**What it does:** Edits text by searching, replacing, inserting, or deleting lines — without opening a text editor.
**Syntax:** `sed [options] 'command' filename`
**Used for:** Find-and-replace in files, automated text editing in scripts.

```python
%%bash
# Create a sample file
echo -e "Hello World\nHello Linux\nHello Students\nLinux is great" > /tmp/sed_demo.txt

echo "=== sed 's/old/new/' — Substitute (replace) first occurrence per line ==="
# s = substitute command
# s/Hello/Hi/ means: find 'Hello', replace with 'Hi' (first occurrence per line)
sed 's/Hello/Hi/' /tmp/sed_demo.txt

echo "=== sed 's/old/new/g' — Replace ALL occurrences (g = global) ==="
# The 'g' flag at the end replaces EVERY match on each line, not just the first
sed 's/l/L/g' /tmp/sed_demo.txt

echo "=== sed '/pattern/d' — Delete lines matching pattern ==="
# /Linux/d means: delete every line that contains 'Linux'
sed '/Linux/d' /tmp/sed_demo.txt

echo "=== sed -n '2p' — Print only a specific line number ==="
# -n suppresses automatic printing
# '2p' means: print line 2
sed -n '2p' /tmp/sed_demo.txt

echo "=== sed -i (edit file IN PLACE) ==="
# -i makes changes directly to the file (without this, output only shows on screen)
# ⚠️ This actually modifies the file!
cp /tmp/sed_demo.txt /tmp/sed_backup.txt   # backup first!
sed -i 's/Hello/Greetings/' /tmp/sed_demo.txt
echo "After in-place edit:"
cat /tmp/sed_demo.txt
```

### `awk` — Text Processing Language
**What it does:** Processes text files column by column — like a mini programming language for data.
**Syntax:** `awk 'pattern { action }' filename`
**Used for:** Extracting columns, computing totals, formatting reports.

```python
%%bash
# Create a data file (CSV-like with spaces)
cat << 'EOF' > /tmp/grades.txt
Alice Mathematics 88
Bob Science 72
Carol Mathematics 95
David Science 81
Eve Arts 67
EOF

echo "=== Print specific columns ==="
# $1 = first column (field), $2 = second, $3 = third
# $0 = the entire line
# awk reads each line and applies the action inside { }
awk '{ print $1, $3 }' /tmp/grades.txt   # print name and score

echo "=== Add custom labels to output ==="
# You can mix your own text with column values
awk '{ print "Student:", $1, "| Subject:", $2, "| Score:", $3 }' /tmp/grades.txt

echo "=== Filter rows based on condition ==="
# If $3 (the score) is greater than 80, print the student name and score
awk '$3 > 80 { print $1, "PASSED with score", $3 }' /tmp/grades.txt

echo "=== Calculate average score ==="
# sum += $3 adds each score to a running total
# count++ increments a counter
# END block runs AFTER all lines are processed
awk '{ sum += $3; count++ } END { printf "Average score: %.1f\n", sum/count }' /tmp/grades.txt

echo "=== Count rows per subject ==="
# NR = current line number, NF = number of fields on this line
awk '{ subjects[$2]++ } END { for (s in subjects) print s, subjects[s], "students" }' /tmp/grades.txt
```

## 🔗 Pipes, Redirection & `xargs`

### Output Redirection (`>`, `>>`, `2>`, `|`)
**What it does:** Controls where command output goes — screen, file, or another command.
**Used for:** Saving output, chaining commands, handling errors separately.

```python
%%bash
echo "=== > (overwrite) ==="
# > redirects output INTO a file, REPLACING its existing contents
echo "First entry" > /tmp/log.txt
echo "Second entry" > /tmp/log.txt   # this REPLACES 'First entry'
cat /tmp/log.txt

echo "=== >> (append) ==="
# >> ADDS to the end of a file without erasing what's already there
echo "Line 1" > /tmp/append.txt
echo "Line 2" >> /tmp/append.txt
echo "Line 3" >> /tmp/append.txt
cat /tmp/append.txt

echo "=== 2> (redirect errors) ==="
# 2 means 'file descriptor 2' which is the error stream (stderr)
# This captures error messages into a file instead of showing them on screen
ls /nonexistent_folder 2> /tmp/errors.txt
echo "Error captured:"; cat /tmp/errors.txt

echo "=== | (pipe) — chain commands together ==="
# The pipe | takes the OUTPUT of the left command and feeds it as INPUT to the right
# Here we list all files in /etc, then filter to show only those with 'host'
ls /etc | grep "host"

echo "=== Multiple pipes (pipeline) ==="
# You can chain multiple commands — each pipes into the next
# ls /etc lists files, grep filters for 'conf', sort alphabetizes, head takes first 5
ls /etc | grep "conf" | sort | head -5
```

### `xargs` — Build Commands from Input
**What it does:** Takes lines of input and passes them as arguments to another command.
**Syntax:** `command | xargs [options] command`
**Used for:** Running commands on a list of files, parallel execution.

```python
%%bash
echo "=== Basic xargs usage ==="
# echo outputs three words, xargs passes each as argument to echo
echo "apple banana cherry" | xargs echo "Fruits:"

echo "=== xargs -n 1 (one argument per command) ==="
# -n 1 means run the command once PER argument
echo "apple banana cherry" | xargs -n 1 echo "Item:"

echo "=== xargs with find (delete files) ==="
# find gives a list of files, xargs passes them to rm for deletion
touch /tmp/xargs1.tmp /tmp/xargs2.tmp /tmp/xargs3.tmp
echo "Before: $(ls /tmp/*.tmp)"
find /tmp -name "*.tmp" | xargs rm
echo "After: all .tmp files deleted"

echo "=== xargs with -I (place argument at specific position) ==="
# -I {} means: put the argument where {} appears in the command
echo -e "Alice\nBob\nCarol" | xargs -I {} echo "Welcome, {}!"
```

## 🖇️ File Links

### `ln` — Create Links Between Files
**What it does:** Creates hard links or symbolic (soft) links to files.
**Syntax:** `ln [options] source link_name`
**Used for:** Creating shortcuts, sharing files across directories.

```python
%%bash
echo "Original content" > /tmp/source_file.txt

echo "=== Hard Link ==="
# ln source linkname — creates a HARD link
# A hard link is another name for the SAME file data
# Both names point to the exact same content on disk (same inode number)
ln /tmp/source_file.txt /tmp/hardlink.txt

# ls -li shows the inode number (first column) — hard links share the same inode
ls -li /tmp/source_file.txt /tmp/hardlink.txt
echo "Both files have the SAME inode number — they are the same file!"

echo "=== Symbolic (Soft) Link ==="
# ln -s source linkname — creates a SYMBOLIC (shortcut) link
# A symlink is like a desktop shortcut — it POINTS to the original file
# If you delete the original, the symlink breaks
ln -s /tmp/source_file.txt /tmp/symlink.txt

# The -> shows where the symlink points
ls -la /tmp/symlink.txt
echo "Reading through symlink:"
cat /tmp/symlink.txt   # reads the original file's content via the link

echo ""
echo "Hard link vs Symbolic link:"
echo "  Hard link   = another name for the SAME data (same inode)"
echo "  Soft link   = a SHORTCUT that points to another file"
```

## 🔁 Synchronization & Transfer

### `rsync` — Remote/Local File Sync
**What it does:** Syncs files between two locations, only copying what changed (very efficient).
**Syntax:** `rsync [options] source/ destination/`
**Used for:** Backups, deploying files, mirroring directories.

```python
%%bash
# Setup source directory with some files
mkdir -p /tmp/rsync_src
echo "Document A" > /tmp/rsync_src/docA.txt
echo "Document B" > /tmp/rsync_src/docB.txt
mkdir -p /tmp/rsync_dst

echo "=== rsync -av (archive + verbose) ==="
# -a = archive mode: preserves permissions, timestamps, symlinks
# -v = verbose: prints the name of each file being copied
rsync -av /tmp/rsync_src/ /tmp/rsync_dst/

echo "Destination contents:"
ls /tmp/rsync_dst/

echo "=== Run rsync again — only CHANGED files are copied ==="
# rsync is smart: if files already exist and haven't changed, they are SKIPPED
echo "Modified doc" > /tmp/rsync_src/docA.txt   # change one file
rsync -av /tmp/rsync_src/ /tmp/rsync_dst/
# Only docA.txt was copied because it was the only changed file

echo "=== rsync --dry-run — preview what WOULD happen without actually copying ==="
# --dry-run is a safety feature: shows what rsync would do without doing it
rsync -av --dry-run /tmp/rsync_src/ /tmp/rsync_dst/
```

## 📜 Shell Scripting

### `bash` — Write and Run Shell Scripts
**What it does:** Executes a file of shell commands as a program (script).
**Syntax:** `bash script.sh`
**Used for:** Automating repetitive tasks, writing programs in shell.

```python
%%bash
# Write a complete shell script that demonstrates variables, loops, and conditions
cat << 'SCRIPT' > /tmp/demo_script.sh
#!/bin/bash
# The shebang line #!/bin/bash tells the OS to use bash to run this script

# --- VARIABLES ---
STUDENT_NAME="Alice"      # define a variable (no spaces around =)
PASS_MARK=50              # numeric variable

echo "=== Variables ==="
echo "Student: $STUDENT_NAME"     # use $ to access a variable's value
echo "Pass mark: $PASS_MARK"

# --- IF / ELSE CONDITION ---
echo ""
echo "=== Conditions ==="
SCORE=75
if [ $SCORE -ge $PASS_MARK ]; then   # -ge means 'greater than or equal to'
    echo "$STUDENT_NAME scored $SCORE — PASSED!"
else
    echo "$STUDENT_NAME scored $SCORE — FAILED."
fi    # fi closes the if block

# --- FOR LOOP ---
echo ""
echo "=== For Loop ==="
for FRUIT in Apple Banana Cherry; do   # loop through a list of items
    echo "  Fruit: $FRUIT"             # $FRUIT changes each iteration
done    # done closes the loop

# --- WHILE LOOP ---
echo ""
echo "=== While Loop ==="
COUNT=1
while [ $COUNT -le 3 ]; do    # keep looping while COUNT is <= 3
    echo "  Count is: $COUNT"
    COUNT=$((COUNT + 1))      # $(( )) is used for arithmetic
done

echo ""
echo "Script complete!"
SCRIPT

# Make the script executable
chmod +x /tmp/demo_script.sh

# Run it!
bash /tmp/demo_script.sh
```

```python
%%bash
# ============================================
# FINAL CHALLENGE — Student Grade Report
# Uses: cat, echo, grep, sort, awk, wc, diff
# ============================================

# Step 1: Create the student data file
cat << 'EOF' > /tmp/class_grades.txt
Alice Science 88
Bob Mathematics 55
Carol Science 92
David Mathematics 73
Eve Science 45
Frank Mathematics 98
Grace Science 61
Henry Mathematics 80
EOF

echo "======================================"
echo "       STUDENT GRADE REPORT           "
echo "======================================"

# Step 2: Count total students using wc -l
echo "Total students: $(wc -l < /tmp/class_grades.txt)"

# Step 3: Use grep to filter students by subject
echo ""
echo "--- Science Students ---"
grep "Science" /tmp/class_grades.txt

# Step 4: Use awk to find students who passed (score >= 60)
echo ""
echo "--- PASSED (score >= 60) ---"
awk '$3 >= 60 { print $1, $3 }' /tmp/class_grades.txt | sort -k2 -rn

# Step 5: Use awk to find students who failed
echo ""
echo "--- FAILED (score < 60) ---"
awk '$3 < 60 { print $1, $3 }' /tmp/class_grades.txt

# Step 6: Use sort + tail to find top scorer
echo ""
echo "--- TOP SCORER ---"
sort -k3 -n /tmp/class_grades.txt | tail -1 | awk '{ print $1, "scored", $3 }'

# Step 7: Use awk to compute the class average
echo ""
echo "--- CLASS AVERAGE ---"
awk '{ sum += $3; count++ } END { printf "Average: %.1f / 100\n", sum/count }' /tmp/class_grades.txt

# Step 8: Save the report to a file
echo ""
echo "Saving report to /tmp/grade_report.txt..."
echo "Report generated on: $(date)" > /tmp/grade_report.txt
awk '{ print $1, $2, $3 }' /tmp/class_grades.txt | sort -k3 -rn >> /tmp/grade_report.txt
echo "Done! Report saved."
cat /tmp/grade_report.txt
```
